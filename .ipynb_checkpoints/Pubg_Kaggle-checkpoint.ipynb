```python
import pandas as pd
```

```python
train = pd.read_csv('train_V2.csv')
```

```python
train
```

```python
train.hist(figsize = (15, 15))
```

```python
train_solo = train[train['matchType'] == 'solo']
```

```python
train_solo2 = train[train['matchType'] == 'solo-fpp']
```

```python
train_solo_merge = pd.concat([train_solo, train_solo2], axis=0)
```

```python
train_solo_merge.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(30,30))
hm = sns.heatmap(train_solo_merge.corr(), annot=True, annot_kws={'size':14})
bottom, top = hm.get_ylim()
hm.set_ylim(bottom+0.5, top-0.5)
plt.tight_layout()
plt.show()
```
